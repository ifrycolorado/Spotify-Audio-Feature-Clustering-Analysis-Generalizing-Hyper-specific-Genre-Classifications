```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
scores_df = pd.read_csv("../data/kmeans/scaled/silhouette_scores.csv", index_col = 0)
display(scores_df)
```

```python
plt.style.use('bmh')
plt.rcParams['figure.figsize'] = (18, 9)

font = {'family': 'sans-serif', 
        'style': 'normal', 
        'weight': 'bold', 
        'size': 10}
plt.rc('font', **font)

spot_green = "#1db954"
spot_white = "#ffffff"
spot_black = "#191414"
```

```python


for i in range(len(scores_df)):

    name = scores_df.iloc[i]["name"]
    
    name_scores = scores_df.loc[scores_df["name"] == name].values.flatten().tolist()
    
    plt.plot([x for x in range(2,21)], name_scores[1:])

plt.ylabel("Silhouette Scores")
plt.xlabel("K")
plt.xticks(ticks=[x for x in range(2,21)], labels=[x for x in range(2,21)])
plt.title("Sihouette Scores v. K")
plt.legend(scores_df["name"])

plt.savefig("../data/kmeans/scaled/images/silhouette_scores.eps", format = 'eps')
plt.savefig("../data/kmeans/scaled/images/silhouette_scores.png", format = 'png')


plt.show()
```
